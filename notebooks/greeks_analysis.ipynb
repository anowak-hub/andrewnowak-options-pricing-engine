```python
# -- Delta vs Stock Price Plot
# Only Focused on Visualizing General Option Sensitivity

# Import
import numpy as np 
import matplotlib.pyplot as plt 
import sys 
import os 
sys.path.append(os.path.abspath(".."))

from src.greeks import (
    call_delta,
    gamma,
    vega,
    call_theta
)

# Create Stock Price Range
stock_prices = np.linspace(50, 150, 100)

# Compute Delta Values
deltas = []

for S in stock_prices:
    delta = call_delta(S, 100, 1, 0.05, 0.2)
    deltas.append(delta)

# Plot the Data 
plt.plot(stock_prices, deltas)
plt.title("Call Delta vs Stock Price")
plt.xlabel("Stock Price")
plt.ylabel("Delta")

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# -- Gamma vs Stock Price Plot

# Import
import numpy as np 
import matplotlib.pyplot as plt 
import sys 
import os 
sys.path.append(os.path.abspath(".."))

from src.greeks import (
    call_delta,
    gamma,
    vega,
    call_theta
)

# Create Stock Price Range
stock_prices = np.linspace(50, 150, 100)

# Compute Gamma Values 
gammas = []

for S in stock_prices:
    gammas.append(gamma(S, 100, 1, 0.05, 0.2))

# Plot the Data 
plt.plot(stock_prices, gammas)
plt.title("Gamma vs Stock Price")
plt.xlabel("Stock Price")
plt.ylabel("Gamma")

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# -- Vega vs Expiration Plot
 
# Import
import numpy as np 
import matplotlib.pyplot as plt 
import sys 
import os 
sys.path.append(os.path.abspath(".."))

from src.greeks import (
    call_delta,
    gamma,
    vega,
    call_theta
)

# Create Expiration Range 
times = np.linspace(0.01, 2, 100)

# Compute Vegas
vegas = []

for T in times:
    vegas.append(vega(100, 100, T, 0.05, 0.2))

# Plot the Data
plt.plot(times, vegas)
plt.title("Vega vs Time to Expiration")
plt.xlabel("Time to Expiration (Years)")
plt.ylabel("Vega")

plt.grid(True)
plt.tight_layout()
plt.show()
```
